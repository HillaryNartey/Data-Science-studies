```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as soup
```

```python
url = 'https://www.jumia.com.gh/phones-tablets/'
url
```

```python
request = Request(url,headers={'User-Agent': 'Mozilla/5.0'})
request
```

```python
webpage = urlopen(request).read()
```

```python
formatted_webpage = soup(webpage, "html.parser")
```

```python
results = [pro.get_text() for pro in formatted_webpage .select(".core .name")]
results
```

```python
price = [cost.get_text() for cost in formatted_webpage .select(".core .prc")]
price

```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "name" : results,
    "Price" : price,})
df
```

```python
df.to_csv('Assignment_results.csv')
```
